```python
reason_val = df['title'].apply(lambda x : x.split(':')[0])
df['Reason'] = reason_val
```

```python
df['Reason'].value_counts().head(5)
```

```python
df
```

```python
sns.countplot(x = 'Reason' , data = df)
```

```python
type(df['timeStamp'].iloc[0])
```

```python

df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
type(df['timeStamp'].iloc[0])
```

```python
time = df['timeStamp'].iloc[0]
hours = time.hour
minute = time.minute
sec = time.second
df['timeStamp'].iloc[12].month

```

```python
df['hours'] = df['timeStamp'].apply(lambda x : x.hour)
df['minutes'] = df['timeStamp'].apply(lambda x : x.minute)
df['seconds'] = df['timeStamp'].apply(lambda x : x.second)
df['Days of Week'] = df['timeStamp'].apply(lambda x : x.dayofweek)
df['Month'] = df['timeStamp'].apply(lambda x : x.month)
```

```python
df.head()
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
dmap
```

```python
df['Days of Week'] = df['Days of Week'].map(dmap)
```

```python
sns.countplot('Days of Week' , hue = 'Reason' , data = df , palette = 'viridis')
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0)
```

** Now use seaborn to create a countplot of the Day of Week column with the hue based off of the Reason column. **

```python
sns.countplot(x = 'Month' , hue = 'Reason' , data = df , palette = 'viridis')
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.0)
```

```python
df['Date'] = df['timeStamp'].apply(lambda x : x.date())
```

```python
df.head()
```

```python
byDate = df.groupby(by = ['Date']).count()
byDate
```

```python
byDate['lat'].plot().grid()
plt.tight_layout()
```

```python
df[df['Reason'] == 'Traffic'].groupby(['Date'])['lat'].count().plot().grid()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason'] == 'Fire'].groupby(['Date'])['lat'].count().plot().grid()
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reason'] == 'EMS'].groupby(['Date'])['lat'].count().plot().grid()


plt.title('EMS')
plt.tight_layout()
```
